#1 Explain the difference between Relative , Absolute and Mixed Cell Referencing.You have the following dataset.

    -In Excel, cell referencing determines how cell addresses behave when a formula is copied or moved to another cell. There are three main types of cell references:

    A. Relative Cell Reference

A relative reference changes automatically when the formula is copied to another cell.

It adjusts both the row and column according to the new position.

Example:
If I write =A1 + B1 in cell C1 and copy it to C2, it becomes =A2 + B2


  B. Absolute Cell Reference

An absolute reference does not change when copied or moved.

It is locked by adding a $ symbol before the column letter and row number.
Abslute cell refrence is also work with function4 (f4) button

Example:
$A$1 always refers to cell A1, no matter where the formula is copied.

  C. Mixed Cell Reference

A mixed reference locks either the row or the column, but not both.

Two possible forms:

$A1 → column fixed, row changes

A$1 → row fixed, column changes

Example:
If =$A1 + B$2 is copied to another cell, $A1 will keep column A fixed, but the row changes; B$2 keeps row 2 fixed but allows the column to move.

*Use it when: I need a formula that partly changes and partly stays fixed.
